
# Q3(ii)

We apply Newton's method with initial guess $x_0 = \tfrac12$ to approximate the root $\alpha = 0$ of:

- (a) $f(x) = \sin x\, \sinh x$
- (b) $f(x) = x - \sin x$
- (c) $f(x) = x \cos x$

In each case we stop when $|f(x_k)| < 10^{-8}$.

The previous question showed two key facts:

- if $f'(\alpha) \neq 0$, Newton's method is at least quadratically convergent;
- if $f'(\alpha) = 0$ but $f''(\alpha) \neq 0$, Newton's method is linearly convergent with rate $\tfrac12$.

These examples illustrate both behaviours, and also show what happens for a higher-multiplicity root.



## The Newton iteration we will use

We use exactly the implementation below. Here `g` denotes $f'$, the derivative of $f$.

The update is
$$
x_{k+1} = x_k - \frac{f(x_k)}{g(x_k)}.
$$

The printed output lets us see the iterates directly.


````python

from math import *

def newton(f, g, x0, tol):
    # Initial evaluation
    x = x0
    fx = f(x)
    iterations = 0
    print(f"Iteration {iterations:2d}: x = {x:e}  f(x) = {fx:e}")
    while abs(fx) > tol:
        iterations += 1
        x = x - fx / g(x)
        fx = f(x)
        print(f"Iteration {iterations:2d}: x = {x:e}  f(x) = {fx:e}")
    return x
````


## A note on `lambda`

In Python, a `lambda` function is a short anonymous function written in one line.

For example,

```python
f = lambda x: x - sin(x)
```

means the same as

```python
def f(x):
    return x - sin(x)
```

Using `lambda` is elegant here because:

- it keeps the function definition next to the Newton call;
- it mirrors the mathematical formula very closely;
- it avoids creating several separate named functions when each formula is only used once.

So for a short computational exercise, `lambda` keeps the code compact and readable.



## Part (a): $f(x) = \sin x \, \sinh x$

Near $x=0$,
$$
\sin x \sim x, \qquad \sinh x \sim x,
$$
so
$$
f(x) \sim x^2.
$$
Thus $x=0$ is a double root. From the previous question, Newton's method should therefore be **linearly convergent** with asymptotic factor $\tfrac12$.


````python

print("Part (a)")
print("-" * 80)
f = lambda x: sin(x) * sinh(x)
g = lambda x: cos(x) * sinh(x) + sin(x) * cosh(x)
root_a = newton(f, g, 0.5, 1e-8)
print(f"Approximate root = {root_a:.16e}")
````


The iterates decrease by essentially a factor of $\tfrac12$ each step once they are close to the root, matching the theory for a double root.



## Part (b): $f(x) = x - \sin x$

Using the Taylor expansion of $\sin x$,
$$
\sin x = x - \frac{x^3}{6} + O(x^5),
$$
so
$$
x - \sin x = \frac{x^3}{6} + O(x^5).
$$
Thus $x=0$ is a **triple root**.

For a root of multiplicity $m$, standard Newton iteration converges only linearly, with factor
$$
\frac{m-1}{m}.
$$
Here $m=3$, so we expect a linear factor of
$$
\frac{2}{3}.
$$


````python

print("Part (b)")
print("-" * 80)
f = lambda x: x - sin(x)
g = lambda x: 1 - cos(x)
root_b = newton(f, g, 0.5, 1e-8)
print(f"Approximate root = {root_b:.16e}")
````


The printed iterates show slower linear convergence than in part (a): the error is multiplied by about $\tfrac23$ each step, consistent with a triple root.



## Part (c): $f(x) = x \cos x$

Here
$$
f(0) = 0,
$$
and
$$
f'(x) = \cos x - x \sin x,
$$
so
$$
f'(0) = 1 \neq 0.
$$
Thus $x=0$ is a **simple root**. From the previous question, Newton's method should therefore converge **quadratically**.


````python

print("Part (c)")
print("-" * 80)
f = lambda x: x * cos(x)
g = lambda x: cos(x) - x * sin(x)
root_c = newton(f, g, 0.5, 1e-8)
print(f"Approximate root = {root_c:.16e}")
````


The iterates collapse to zero very quickly. This is the expected quadratic convergence for a simple root: once $x_k$ is small, the next error is proportional to $|x_k|^2$.



## Summary

- **(a)** $f(x) = \sin x\, \sinh x$ has a double root at $0$, so Newton's method is **linear** with rate $\tfrac12$.
- **(b)** $f(x) = x - \sin x$ has a triple root at $0$, so Newton's method is **linear** with rate $\tfrac23$.
- **(c)** $f(x) = x \cos x$ has a simple root at $0$, so Newton's method is **quadratic**.

So the observed behaviour agrees with the convergence theory: simple roots give quadratic convergence, while multiple roots reduce Newton's method to linear convergence.
